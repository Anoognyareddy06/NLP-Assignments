**Next Word Prediction Using Tri-Gram Counts**

```python
def predict_next_word_trigram(word_sequence, Trigrams_counts, bigram_counts):
    # Tokenize the input sequence
    words_in_sequence = word_sequence.lower().split()
    if not words_in_sequence:
        return "Please provide a word sequence."

    # Ensure at least two words for trigram prediction
    if len(words_in_sequence) < 2:
        return "Sequence must contain at least two words for trigram prediction."

    # Get the last two words as a tuple
    last_two_words_tuple = tuple(words_in_sequence[-2:])

    # Find potential next words based on trigrams starting with the last two words
    potential_next_words = {}
    for (w1, w2, w3), count in Trigrams_counts.items():
        if (w1, w2) == last_two_words_tuple:
            potential_next_words[w3] = count

    if not potential_next_words:
        return f"No trigram found starting with '{' '.join(last_two_words_tuple)}'."

    # Calculate probabilities for potential next words
    # P(w3 | w1,w2) = Count(w1, w2, w3) / Count(w1, w2)
    # The denominator should be the count of the bigram (w1, w2)
    last_two_words_bigram_count = bigram_counts.get(last_two_words_tuple, 0)
    if last_two_words_bigram_count == 0:
        return f"'{' '.join(last_two_words_tuple)}' not found as a bigram. Cannot predict next word."

    predicted_word = None
    max_probability = -1

    for next_word, trigram_count in potential_next_words.items():
        probability = trigram_count / last_two_words_bigram_count
        print("probability of  " f'{next_word}' " is ", probability)
        if probability > max_probability:
            max_probability = probability
            predicted_word = next_word

    return predicted_word

# Example usage:
# Ensure bigram_counts, unigram_counts, and Trigrams_counts are defined from previous cells
# (They are present in the kernel state.)

# Try with a word sequence


sequence1 = "I am pursuing"
print(f"Given sequence: '{sequence1}', predicted next word: '{predict_next_word_trigram(sequence1, Trigrams_counts, bigram_counts)}'")

sequence2 = "my engineering course"
print(f"Given sequence: '{sequence2}', predicted next word: '{predict_next_word_trigram(sequence2, Trigrams_counts, bigram_counts)}'")

sequence3 = "my intermediate education"
print(f"Given sequence: '{sequence3}', predicted next word: '{predict_next_word_trigram(sequence3, Trigrams_counts, bigram_counts)}'")

sequence4 = "a software developer"
print(f"Given sequence: '{sequence4}', predicted next word: '{predict_next_word_trigram(sequence4, Trigrams_counts, bigram_counts)}'")

sequence5 = "artificial intelligence and"
print(f"Given sequence: '{sequence5}', predicted next word: '{predict_next_word_trigram(sequence5, Trigrams_counts, bigram_counts)}'")

sequence6 = "coding contests to"
print(f"Given sequence: '{sequence6}', predicted next word: '{predict_next_word_trigram(sequence6, Trigrams_counts, bigram_counts)}'")

sequence7 = "web applications using"
print(f"Given sequence: '{sequence7}', predicted next word: '{predict_next_word_trigram(sequence7, Trigrams_counts, bigram_counts)}'")

sequence8 = "cloud computing and"
print(f"Given sequence: '{sequence8}', predicted next word: '{predict_next_word_trigram(sequence8, Trigrams_counts, bigram_counts)}'")

sequence9 = "data analytics and"
print(f"Given sequence: '{sequence9}', predicted next word: '{predict_next_word_trigram(sequence9, Trigrams_counts, bigram_counts)}'")

sequence10 = "innovative technology solutions"
print(f"Given sequence: '{sequence10}', predicted next word: '{predict_next_word_trigram(sequence10, Trigrams_counts, bigram_counts)}'")
```

**Next Word Prediction Using Tri-Gram Counts with Add - K Smoothening**

```python
def predict_next_word_trigram_K(word_sequence, Trigrams_counts, bigram_counts,K): #K=0.5-0.01
    # Tokenize the input sequence
    words_in_sequence = word_sequence.lower().split()
    if not words_in_sequence:
        return "Please provide a word sequence."

    # Ensure at least two words for trigram prediction
    if len(words_in_sequence) < 2:
        return "Sequence must contain at least two words for trigram prediction."

    # Get the last two words as a tuple
    last_two_words_tuple = tuple(words_in_sequence[-2:])

    # Find potential next words based on trigrams starting with the last two words
    potential_next_words = {}
    for (w1, w2, w3), count in Trigrams_counts.items():
        if (w1, w2) == last_two_words_tuple:
            potential_next_words[w3] = count

    if not potential_next_words:
        return f"No trigram found starting with '{' '.join(last_two_words_tuple)}'."

    # Calculate probabilities for potential next words
    # P(w3 | w1,w2) = Count(w1, w2, w3) / Count(w1, w2)
    # The denominator should be the count of the bigram (w1, w2)
    last_two_words_bigram_count = bigram_counts.get(last_two_words_tuple, 0)
    if last_two_words_bigram_count == 0:
        return f"'{' '.join(last_two_words_tuple)}' not found as a bigram. Cannot predict next word."

    predicted_word = None
    max_probability = -1

    for next_word, trigram_count in potential_next_words.items():
        probability = (trigram_count+K) / (last_two_words_bigram_count+K*V)
        print("probability of  " f'{next_word}' " is ", probability)
        if probability > max_probability:
            max_probability = probability
            predicted_word = next_word

    return predicted_word

# Example usage:
# Ensure bigram_counts, unigram_counts, and Trigrams_counts are defined from previous cells
# (They are present in the kernel state.)

# Try with a word sequence
sequence1 = "I am pursuing"
next_word1 = predict_next_word_trigram_K(sequence1, Trigrams_counts, bigram_counts, K=0.5)
print(f"Given sequence: '{sequence1}', predicted next word: '{next_word1}'")

sequence2 = "During my engineering"
next_word2 = predict_next_word_trigram_K(sequence2, Trigrams_counts, bigram_counts, K=0.5)
print(f"Given sequence: '{sequence2}', predicted next word: '{next_word2}'")

sequence3 = "I completed my"
next_word3 = predict_next_word_trigram_K(sequence3, Trigrams_counts, bigram_counts, K=0.5)
print(f"Given sequence: '{sequence3}', predicted next word: '{next_word3}'")

sequence4 = "My ambition is"
next_word4 = predict_next_word_trigram_K(sequence4, Trigrams_counts, bigram_counts, K=0.5)
print(f"Given sequence: '{sequence4}', predicted next word: '{next_word4}'")

sequence5 = "I am interested"
next_word5 = predict_next_word_trigram_K(sequence5, Trigrams_counts, bigram_counts, K=0.5)
print(f"Given sequence: '{sequence5}', predicted next word: '{next_word5}'")

sequence6 = "I actively participate"
next_word6 = predict_next_word_trigram_K(sequence6, Trigrams_counts, bigram_counts, K=0.5)
print(f"Given sequence: '{sequence6}', predicted next word: '{next_word6}'")

sequence7 = "I have developed"
next_word7 = predict_next_word_trigram_K(sequence7, Trigrams_counts, bigram_counts, K=0.5)
print(f"Given sequence: '{sequence7}', predicted next word: '{next_word7}'")

sequence8 = "I want to"
next_word8 = predict_next_word_trigram_K(sequence8, Trigrams_counts, bigram_counts, K=0.5)
print(f"Given sequence: '{sequence8}', predicted next word: '{next_word8}'")

sequence9 = "I am passionate"
next_word9 = predict_next_word_trigram_K(sequence9, Trigrams_counts, bigram_counts, K=0.5)
print(f"Given sequence: '{sequence9}', predicted next word: '{next_word9}'")

sequence10 = "My long-term goal"
next_word10 = predict_next_word_trigram_K(sequence10, Trigrams_counts, bigram_counts, K=0.5)
print(f"Given sequence: '{sequence10}', predicted next word: '{next_word10}'")
```
